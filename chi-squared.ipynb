The chi2_contingency() function of scipy.stats module takes the contingency table element and it returns a tuple containing test statistics,  p-value, degrees of freedom, and expected table(the one we created from the calculated values) in that order. 

```python
from scipy.stats import chi2_contingency

# defining the table
data = [[6,7], [13,16], [16,15], [8,11]]
stat, p, dof, expected = chi2_contingency(data)

print("chi-square =", stat)
print("Degree of Freedom = ", dof)
# print("p value = " , p)
# print("", expected)

```
